```python
import pandas as pd
import os

def load_titanic_train_dataset(dataset_path):
  csv_path = os.path.join(dataset_path, "train.csv")
  return pd.read_csv(csv_path)

def load_titanic_test_dataset(dataset_path):
  csv_path = os.path.join(dataset_path, "test.csv")
  return pd.read_csv(csv_path)
```

```python
dir_path = "/content/titanic"

train_set = load_titanic_train_dataset(dir_path)
test_set = load_titanic_test_dataset(dir_path)
```

# Get data description. Clean the data

```python
train_set.head()
```

```python
train_set.describe()
```

Train set is missing age value for 177 passengers. Lets fill the missing data with age median.

```python
age_median = train_set["Age"].median()
train_set["Age"].fillna(age_median, inplace=True)
train_set.describe()
```

```python
test_set.head()
```

```python
test_set.describe()
```

Test set has the same problem. It has 86 missing values for age feature and 1 missing value for fare feature. Lets apply the same technique as for train set replacing empty values with median.

```python
test_age_median = test_set["Age"].median()
test_set["Age"].fillna(test_age_median, inplace=True)
test_fare_median = test_set["Fare"].median()
test_set["Fare"].fillna(test_fare_median, inplace=True)
test_set.describe()
```

Perfect. No numerical values are missing in both train and test sets. The other part of data cleaning will be related to text values. I think that no real value comes from *Name* column so there should be the possibility to drop this column. The *Sex* column will be binary encoded as 0 - male and 1 for female. *Ticket* and *Cabin* columns will also be dropped.
